# Import

```python
import pandas as pd
import os
from datetime import datetime
import re
```

# Création classement

```python
data = {
    'Team': [
        "ATL", "BOS", "BRK", "CHO", "CHI", "CLE", "DAL", "DEN", "DET", "GSW", "HOU", "IND", "LAC", "LAL", "MEM", 
        "MIA", "MIL", "MIN", "NOP", "NYK", "OKC", "ORL", "PHI", "PHO", "POR", "SAC", "SAS", "TOR", "UTA", "WAS"
    ],
    'Match': [0 for _ in range(30)],
    'Win': [0 for _ in range(30)]
}

df_info_match = pd.DataFrame(data)
```

```python
df_info_match.head()
```

```python
colonnes = ['Team', 'Match', 'Win', 'Date']
df_classement = pd.DataFrame(columns=colonnes)
```

```python
dossier_box_score = "data_2023_2024/box_score_regular_season"

for dir in os.listdir(dossier_box_score):
    match = re.match(r'box_sore_(\d{8})', dir)

    if match:
        date_str = match.group(1) 
    
        # Extraire l'année, le mois et le jour
        year = int(date_str[:4])
        month = int(date_str[4:6])
        day = int(date_str[6:8])

    else:
        print("Le format de la variable ne correspond pas aux attentes.")

    match_dir = os.path.join(dossier_box_score, dir)
    for info in os.listdir(match_dir):
        if info == "df_line_score.csv":

            line_score_path = os.path.join(match_dir, info)
            df_line_score = pd.read_csv(line_score_path)

            team_1, team_2 = df_line_score["Team"].tolist()
            result = df_line_score["Total"].tolist()
            
            df_classement.loc[len(df_classement)] = [team_1, df_info_match[df_info_match["Team"] == team_1]["Match"].values[0],
                                df_info_match[df_info_match["Team"] == team_1]["Win"].values[0], datetime(year, month, day)]

            df_classement.loc[len(df_classement)] = [team_2, df_info_match[df_info_match["Team"] == team_2]["Match"].values[0],
                                df_info_match[df_info_match["Team"] == team_2]["Win"].values[0], datetime(year, month, day)]

            df_info_match.loc[df_info_match['Team'] == team_1, 'Match'] += 1
            df_info_match.loc[df_info_match['Team'] == team_2, 'Match'] += 1

            if result[0] > result[1]:
                df_info_match.loc[df_info_match['Team'] == team_1, 'Win'] += 1
            else:
                df_info_match.loc[df_info_match['Team'] == team_2, 'Win'] += 1
```

```python
df_classement.head()
```

```python
path_to_save = "data_2023_2024/win_per_day.csv"

df_classement.to_csv(path_to_save, index=False)
```
