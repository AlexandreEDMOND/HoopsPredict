# Import

```python
import pandas as pd
import os
from datetime import datetime, timedelta
import re
import warnings
warnings.filterwarnings('ignore')
```

# Cr√©ation Classement

```python
data = {
    'Team': [
        "ATL", "BOS", "BRK", "CHO", "CHI", "CLE", "DAL", "DEN", "DET", "GSW", "HOU", "IND", "LAC", "LAL", "MEM", 
        "MIA", "MIL", "MIN", "NOP", "NYK", "OKC", "ORL", "PHI", "PHO", "POR", "SAC", "SAS", "TOR", "UTA", "WAS"
    ],
    'Win': [0 for _ in range(30)],
    'Date':[datetime(2023,10,24).date() for _ in range(30)]
}

list_df_classement = [pd.DataFrame(data)]
```

```python
list_df_classement[0].head(50)
```

# Import Win Per Day

```python
win_per_day_path = "data_2023_2024/win_per_day.csv"

df_win_per_day = pd.read_csv(win_per_day_path)
```

```python
df_win_per_day.head()
```

```python
grouped = df_win_per_day.groupby('Date')

for date, group in grouped:
    if date != "2023-10-24":

        new_df_classement = list_df_classement[-1].copy()

        new_df_classement["Date"]  = date

        for index, row in group.iterrows():
            # print(f"Ligne {index}:")
            # print(row["Team"])
            new_df_classement.loc[new_df_classement["Team"] == row["Team"], 'Win'] = row["Win"]
            # print()

        list_df_classement.append(new_df_classement)

        # print(f"Date: {date}")
        # print(group)
        # print()
```

```python
for df in list_df_classement:
    df['Ranking'] = df['Win'].rank(ascending=False, method='first')
    print(df.head())
```

```python
list_df_classement[-1].sort_values("Win", ascending=False)
```

# Sauvegarde Dataframe

```python
df_ranking_per_day = pd.concat(list_df_classement, ignore_index=True)
```

```python
df_ranking_per_day.head()
```

```python
save_path = "data_2023_2024/ranking_per_day.csv"

df_ranking_per_day.to_csv(save_path, index=False)
```
