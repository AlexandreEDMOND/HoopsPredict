# Création Liste Lien

```python
path_file = "data_2023_2024/date_team_result/date_teams_result_match_23_24_regular_season.csv"

df = pd.read_csv(path_file)
```

```python
df.head()
```

```python
list_date_dataframe = df['Date'].tolist()
list_team = df['Home/Neutral'].tolist()

print(list_date_dataframe)
print(list_team)
```

```python
def transform_list_date(list_date):
    new_list_date = []

    for date in list_date:
        new_date = date.replace('-', '')
        new_date += '0'
        new_list_date.append(new_date)
    
    return new_list_date
```

```python
list_date_dataframe = transform_list_date(list_date_dataframe)

print(list_date_dataframe)
```

```python
# Dictionnaire de noms d'équipes et leurs abréviations
nba_teams = {
    "Atlanta Hawks": "ATL",
    "Boston Celtics": "BOS",
    "Brooklyn Nets": "BRK",
    "Charlotte Hornets": "CHO",
    "Chicago Bulls": "CHI",
    "Cleveland Cavaliers": "CLE",
    "Dallas Mavericks": "DAL",
    "Denver Nuggets": "DEN",
    "Detroit Pistons": "DET",
    "Golden State Warriors": "GSW",
    "Houston Rockets": "HOU",
    "Indiana Pacers": "IND",
    "Los Angeles Clippers": "LAC",
    "Los Angeles Lakers": "LAL",
    "Memphis Grizzlies": "MEM",
    "Miami Heat": "MIA",
    "Milwaukee Bucks": "MIL",
    "Minnesota Timberwolves": "MIN",
    "New Orleans Pelicans": "NOP",
    "New York Knicks": "NYK",
    "Oklahoma City Thunder": "OKC",
    "Orlando Magic": "ORL",
    "Philadelphia 76ers": "PHI",
    "Phoenix Suns": "PHO",
    "Portland Trail Blazers": "POR",
    "Sacramento Kings": "SAC",
    "San Antonio Spurs": "SAS",
    "Toronto Raptors": "TOR",
    "Utah Jazz": "UTA",
    "Washington Wizards": "WAS"
}

def transform_list_team(list_team):
    new_list_team = []

    for team in list_team:
        new_team = nba_teams[team]
        new_list_team.append(new_team)

    return new_list_team
```

```python
list_team = transform_list_team(list_team)

print(list_team)
```

```python
suffixe_lien = []

for i in range(len(list_date_dataframe)):
    suffixe = list_date_dataframe[i] + list_team[i]
    suffixe_lien.append(suffixe)

print(suffixe_lien)
```

```python
list_lien = []

for suffixe in suffixe_lien:
    lien = f"https://www.basketball-reference.com/boxscores/{suffixe}.html"
    list_lien.append(lien)

print(list_lien)
```

# Scrapping / Prise en main

```python
response = requests.get(list_lien[846], verify=False)
```

```python
# Vérifier si la requête a réussi
if response.status_code == 200:
    # Parser le contenu HTML avec BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    print(soup)
```

# Get Line Score

```python
line_score = []

# Extraire tous les commentaires
for comment in soup.find_all(string=lambda text: isinstance(text, Comment)):
    if '<div class="table_container" id="div_line_score">' in comment:
        
        comment_soup = BeautifulSoup(comment, 'html.parser')
        for tbody in comment_soup.find_all('tbody'):
            for row in tbody:
                if row != "\n":
                    teams_info = []
                    for col in row:
                        if col != "\n":
                            teams_info.append(col.text.strip())
                    line_score.append(teams_info)
        
        break  # Arrête après avoir trouvé le commentaire

print(line_score)

df_line_score = pd.DataFrame(line_score, columns=['Team', 'Q1', 'Q2', 'Q3', 'Q4', 'Total'])
```

# Get Inactive

```python
# Trouver les sections des joueurs inactifs
divs = soup.find_all('div')
for div in divs:
    if "Inactive:" == div.text[:9]:

        inactive_text = div.text.replace('\xa0', ' ')

        away_inactive = []
        home_inactive = []

        away_part = re.search(rf'{away_team} (.*) {home_team}', inactive_text)
        if away_part:
            away_inactive = away_part.group(1).split(', ')
            # Nettoyer les espaces supplémentaires
            away_inactive = [name.strip() for name in away_inactive]

        # Trouver la partie après "DEN"
        home_part = re.search(rf'{home_team} (.*)', inactive_text)
        if home_part:
            home_inactive = home_part.group(1).split(', ')
            # Nettoyer les espaces supplémentaires
            home_inactive = [name.strip() for name in home_inactive]

print(away_inactive)
print(home_inactive)
```
