# Import

```python
import requests
from bs4 import BeautifulSoup
import time
```

# Prise en main

```python
day = 29
month = 12
year = 2023


response = requests.get(f'https://www.basketball-reference.com/boxscores/?month={month}&day={day}&year={year}', verify=False)
# Vérifier si la requête a réussi
if response.status_code == 200:
    # Parser le contenu HTML avec BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extraire et afficher un contenu spécifique, par exemple les scores des matchs
    games = soup.find_all('div', class_='game_summary')
    
    for game in games:
        teams = game.find_all('a')

        team1 = teams[0].text
        team2 = teams[2].text
        
        scores = game.find_all('td', class_='right')
        score1 = scores[0].text
        score2 = scores[2].text
        
        print(f'{team1} {score1} - {team2} {score2}')
else:
    print('Failed to retrieve the page')
```

# Stockage Data

```python
class Match():

    def __init__(self, teamHome, teamAway, pointsHome, pointsAway):
        self.teamHome = teamHome
        self.teamAway = teamAway
        self.pointsHome = pointsHome
        self.pointsAway = pointsAway

        self.winnerHome = pointsHome > pointsAway
    
    def affichage_match(self):
        print(f"{self.teamHome} VS {self.teamAway} | {self.pointsHome}-{self.pointsAway} | Winner : {self.teamHome if self.winnerHome else self.teamAway}")
```

# Scrapping

```python
# Fonction pour récupérer et afficher les scores d'une date donnée
def get_scores(date, list_matches):
    response = requests.get(f'https://www.basketball-reference.com/boxscores/?month={date.month}&day={date.day}&year={date.year}', verify=False)

    nb_matchs = 0

    # Vérifier si la requête a réussi
    if response.status_code == 200:
        # Parser le contenu HTML avec BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extraire et afficher un contenu spécifique, par exemple les scores des matchs
        games = soup.find_all('div', class_='game_summary')
        
        for game in games:
            teams = game.find_all('a')

            team1 = teams[0].text
            team2 = teams[2].text
            
            scores = game.find_all('td', class_='right')
            score1 = scores[0].text
            score2 = scores[2].text
            
            list_matches.append(Match(team1, team2, score1, score2))

            nb_matchs += 1
        
        print(f"Nombre de match le {date.day} {date.month} : {nb_matchs}")
    else:
        print('Failed to retrieve the page')
    
    # Pour éviter le ban
    time.sleep(3)
```

```python
from datetime import datetime, timedelta

# Période de temps
start_date = datetime(2023, 10, 23)
# end_date = datetime(2024, 4, 15)
end_date = datetime(2023, 12, 29)

list_matches = []

# Boucle pour chaque jour de la période
current_date = start_date
while current_date <= end_date:
    get_scores(current_date, list_matches)
    current_date += timedelta(days=1)
```
